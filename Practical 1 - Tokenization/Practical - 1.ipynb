```python
import nltk
```

```python
from nltk.tokenize import WhitespaceTokenizer, WordPunctTokenizer, TreebankWordTokenizer, TweetTokenizer, MWETokenizer
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
sentence = "Hello! I am Tulaja Patil and I am successfully completing my NLP practicals using Python libraries. :) #NLP"
```

```python
print("Original Sentence:", sentence)
print("-" * 50)
```

```python
# PART 1: TOKENIZATION
print("1. TOKENIZATION RESULTS :")
```

```python
# a) Whitespace Tokenizer 
print("Whitespace:", WhitespaceTokenizer().tokenize(sentence), "\n")
```

```python
# b) Punctuation-based 
print("Punctuation-based:", WordPunctTokenizer().tokenize(sentence), "\n")
```

```python
# c) Treebank Tokenizer (standard English tokenization)
print("Treebank:", TreebankWordTokenizer().tokenize(sentence), "\n")
```

```python
# d) Tweet Tokenizer (keeps emojis/smileys and hashtags together)
print("Tweet:", TweetTokenizer().tokenize(sentence), "\n")
```

```python
# e) MWE (Multi-Word Expression) 
mwe_tokenizer = MWETokenizer([('Tulaja', 'Patil')]) 
print("MWE:", mwe_tokenizer.tokenize(sentence.split()), "\n")
```

```python
# PART 2: STEMMING

print("2. STEMMING RESULTS:\n")
```

```python
# 1. Create the stemmers
porter = PorterStemmer()
snowball = SnowballStemmer("english")

# 2. Pick just ONE word to test
word = "successfully"

# 3. Print the results directly!
print("Original Word:", word)
print("Porter Stemmer:", porter.stem(word))
print("Snowball Stemmer:", snowball.stem(word))
```

```python
# PART 3: LEMMATIZATION

print("3. LEMMATIZATION RESULTS:\n")
```

```python
lemmatizer = WordNetLemmatizer()

print(f"Word: 'libraries'  -> Lemmatized: {lemmatizer.lemmatize('libraries', pos='n')}")
print(f"Word: 'completing' -> Lemmatized: {lemmatizer.lemmatize('completing', pos='v')}")
```
