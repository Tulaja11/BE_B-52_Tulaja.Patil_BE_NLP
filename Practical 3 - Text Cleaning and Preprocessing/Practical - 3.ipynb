```python
# Load dataset
data = pickle.load(open("News_dataset.pickle","rb"))
```

```python
data = data.head(5).copy() 
```

```python
print(data.columns)
# select correct columns
text = data['Content']
label = data['Category']
```

```python
# initialize tools
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# function for cleaning
def clean_text(sentence):
    # convert to lowercase
    sentence = sentence.lower()
    
    # remove punctuation & numbers
    sentence = re.sub(r'[^a-zA-Z ]', '', sentence)

    # split words
    words = sentence.split()

    # remove stopwords + lemmatization
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]

    return " ".join(words)
```

```python
# apply cleaning
cleaned_text = text.apply(clean_text)
```

```python
# convert label text into numbers
le = LabelEncoder()
encoded_label = le.fit_transform(label)
```

```python
# TF-IDF representation
tfidf = TfidfVectorizer()
```

```python
X = tfidf.fit_transform(cleaned_text)
```

```python
# save outputs
pickle.dump(X, open("tfidf_output.pkl","wb"))
pickle.dump(encoded_label, open("labels.pkl","wb"))
```
